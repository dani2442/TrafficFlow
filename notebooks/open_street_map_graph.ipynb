```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
```

```python
place_name = "Tres Cantos, Spain"

# Download the street network for the specified place
G = ox.graph_from_place(place_name, network_type='drive')

# Plot the street network
fig, ax = ox.plot_graph(G, node_size=0, edge_linewidth=0.8)

# Optionally, save the graph to a file
ox.save_graphml(G, filepath='san_francisco_road_network.graphml')

# Print basic stats about the graph
stats = ox.basic_stats(G)
print(stats)
```
